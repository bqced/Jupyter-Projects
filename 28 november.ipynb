```python
#Filtering
import pandas as pd

data=pd.read_csv(r"C:\Users\user\Downloads\Sales.csv")
sales_over_160 = data[data['Sales'] > 2060]
print(sales_over_160)
```

```python
#Filtering less than
Less_than = data[data['Sales']<35000]
print(Less_than)
```

```python
Hours = data[(data['Discount'] == 0.04 )]
print(Hours)
```

```python
#Tilde ~ means NOT  
#The & sign means AND 
# The verticle line | means OR
```

```python
#The & sign means AND
the_and_sign=data[(data['Unit_Price']>100)&(data['Order_Quantity']>45)]
print(the_and_sign)
```

```python
#OR SIGN
import pandas as pd

data=pd.read_csv(r"C:\Users\user\Downloads\Sales.csv")

the_OR_sign=data[(data['Unit_Price']>100)|(data['Order_Quantity']>45)]
print(the_OR_sign)
```

```python
# NOT ~ examples
Unit_price_more= data[~(data['Unit_Price']>150)]
print(Unit_price_more)
```

```python
#Inverse of AND statment with NOT (~)
the_AND_sign=data[~((data['Unit_Price']>100)&(data['Order_Quantity']>45))]
print(the_AND_sign)
```

```python
#Filtering by Item 
filtering_by_item= data[data['Order_Priority'].isin(['High','Low'])]
print(filtering_by_item)
```

```python
#Filtering by letters
filtering_by_letters=data[data['Region'].str.contains('w',case=False)]
print(filtering_by_letters)
```

```python
#try and search for Region, all values that don't contain 'W'
import pandas as pd

data=pd.read_csv(r"C:\Users\user\Downloads\Sales.csv")
No_W= data[~(data['Region'].str.contains('w',case=False))]
print(No_W)
```

```python
#Inverse of Orders that are High and Low
import pandas as pd

data=pd.read_csv(r"C:\Users\user\Downloads\Sales.csv")
filtering_by_item= data[~data['Order_Priority'].isin(['High','Low'])]
print(filtering_by_item)
```
