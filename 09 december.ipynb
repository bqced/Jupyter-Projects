```python
import pandas as pd
import matplotlib.pyplot as plt

dcx = pd.read_csv("/Users/user/Downloads/Sales.csv")

dcx_sort = dcx.sort_values(by='Sales', ascending=False).head(25)

y = dcx_sort['Sales'][:25]
x = dcx_sort['Customer_Name'][:25]

plt.bar(x, y, color ='pink', label='sales by customer')
plt.xticks(rotation = 'vertical')
plt.xlabel('customer name')
plt.ylabel('sales')
plt.show()
```
